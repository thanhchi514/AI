```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model

url = 'mtcars.csv'
data = pd.read_csv(url)

X = data.drop(data.columns[[0, 2]], axis=1)
y = data['mpg']

range = np.arange(1.5, 5.5, 0.1)

for degree in np.arange(2, 11):

    X_poly = pd.DataFrame([X['wt'] ** i for i in np.arange(1, degree + 1)]).T

    model = linear_model.LinearRegression()
    model.fit(X_poly, y)

   

    predictor = pd.DataFrame([range ** i for i in np.arange(1, degree + 1)]).T
    y_pred = model.predict(predictor)

    print(f'Intercept của mô hình bậc {degree}: {model.intercept_}')
    print(f'Coefficients của mô hình bậc {degree}: {model.coef_}')
    print(f'R-square của mô hình bậc {degree}: {model.score(X_poly, y)}')

    plt.figure()
    plt.scatter(X['wt'], y, color='red')
    plt.plot(range, y_pred, color='blue')
    plt.title('Polynomial degree: ' + str(degree))
    plt.xlabel('wt')
    plt.ylabel('mpg')
    plt.xlim(0, 9)

plt.show()
```
