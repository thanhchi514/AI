```python
import numpy as np
import skfuzzy as fuzz
from skfuzzy import control as ctrl

# Định nghĩa các biến thuộc tính mờ và không gian giá trị của chúng
weight = ctrl.Antecedent(np.arange(0, 101, 1),'weight')

weight['very_light'] = fuzz.trimf(weight.universe, [0, 0, 30])
weight['light'] = fuzz.trimf(weight.universe, [30, 40, 50])
weight['average'] = fuzz.trimf(weight.universe, [45, 50, 55])
weight['heavy'] = fuzz.trimf(weight.universe, [60, 70, 80])
weight['very_heavy'] = fuzz.trimf(weight.universe, [75, 100, 100])

# Vẽ các hàm thuộc tính mờ lên biểu đồ
weight.view()
```
